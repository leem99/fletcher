# Data Gathering for Project Fletcher

```python
import requests
from fake_useragent import UserAgent
import time
from bs4 import BeautifulSoup

import os
import re
import numpy as np
import pandas as pd

import urllib
import json
```

```python
def random_soup(url):
    us = UserAgent()
    user_agent = {'User-Agent':us.random}
    response = requests.get(url,headers = user_agent)
    page = response.text
    soup = BeautifulSoup(page,"lxml")
    return soup
```

```python
congress_sessions = list(map(str,range(104,116)))
```

```python
save_path = 'laws/'
saved_flist = os.listdir(save_path)
```

Download Text of Documents from U.S Governmental Publishing Office

```python
count = 0 
failed_list = []
keep_going = True

for ix, cs in enumerate(congress_sessions):
    
    keep_going = True

    while keep_going:
    
        count += 1    
        next_link = 'https://www.gpo.gov/fdsys/pkg/PLAW-' + cs + \
            'publ' + str(count)+'/html/PLAW-'+ cs +'publ' + \
            str(count) + '.htm'

        save_name = 'c' + cs + 'publ' + str(count) + '.htm'

        if save_name not in  saved_flist:
            time.sleep(1+np.random.uniform(0,3)) 

            law_soup = random_soup(next_link)
            law_htm = law_soup.prettify("utf-8")

            if law_soup.text.find('Error Detected') != -1:
                count = 0
                keep_going = False
                failed_list.append(next_link)
            else:
                with open(save_path + save_name, "wb") as lfile:
                    lfile.write(law_htm)
                    lfile.close()

                saved_flist.append(save_name)
                print(ix, save_name)
```

Get Law Descriptions from Congress.gov

```python
law_links =[]

for ix, cs in enumerate(congress_sessions):
    # Will stop working at 121st congress
    page = random_soup('https://www.congress.gov/public-laws/'+cs+'th-congress')
    time.sleep(1+np.random.uniform(0,3))
    
    law_links += page.find_all('a',
                  href=re.compile(
                      'https://www.congress.gov/bill/'+cs+'th-congress/*/.*\\d$'))
```

Get voting records 

```python
congress_sessions = list(map(str,range(110,116)))
```

```python
count = 0 
bill_year_list = []
keep_going = True

for ix, cs in enumerate(congress_sessions):
    next_link = 'https://www.govtrack.us/data/congress/' + cs + '/votes/'
    time.sleep(1+np.random.uniform(0,1)) 
    year_soup = random_soup(next_link)
    years = year_soup.find_all('a',href = re.compile('[0-9]'))
    
    print(cs)
    for y in years:
        bill_year_list.append(
            next_link+y.text)
```

```python
bill_links = [ ]
for ix, byl in enumerate(bill_year_list):
    time.sleep(1+0.1*np.random.uniform(0,1)) 
    byl_soup = random_soup(byl)
    
    all_links = byl_soup.find_all('a')

    for link in all_links:
        l_text = link.text
        if l_text != '../':
            bill_links.append(byl+l_text)
```
