```python
import pandas as pd
import numpy as np

import nltk
from bs4 import BeautifulSoup
from collections import OrderedDict
import os
import urllib
```

```python
laws_path = 'laws/'
```

```python
# Get a list of files to convert from html to text
law_files = os.listdir(laws_path)
law_files.sort()
law_files.pop(0)
```

Loop Through Files and make a giant dictionary of files

```python
big_dict = dict()
```

```python
for ix, law_file in enumerate(law_files):

    # Open html file
    with open(laws_path+law_file, "r") as f:
        law_page = f.read()


    law_soup = BeautifulSoup(law_page,'lxml')
    law_text = law_soup.find('pre').get_text()

    k = law_file.replace('.htm','')
    big_dict[k] = law_text

    if (ix % 100)== 0:
        print(str(ix)+ ' ' + law_file)
```

```python
laws_df = pd.DataFrame.from_dict(big_dict,orient = 'index')
laws_df.rename(columns ={0:'text'},inplace=True)
laws_df.to_csv('all_laws.csv')
```

Make a verision of the data frame with paragraphs as rows

```python
big_dict2 = dict()

for ix, k in enumerate(big_dict.keys()):
    law = big_dict[k].split('\n\n') 
    for jx, paragraph in enumerate(law):
        wc = len(paragraph.split())
        if wc >= 50:
            new_k = k + '_p' + str(jx)
            big_dict2[new_k] = paragraph

    
```

```python
laws_paragraphs_df = pd.DataFrame.from_dict(big_dict2,orient = 'index')
laws_paragraphs_df.rename(columns ={0:'text'},inplace=True)
laws_paragraphs_df.to_csv('all_laws_paragraphs.csv')
```
